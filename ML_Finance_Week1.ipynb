```python
import numpy as np
sid = 2421463   
last_two_digits = sid % 100     
if last_two_digits < 10:
    num_elements = last_two_digits + 100
else:
    num_elements = last_two_digits
print("SID:", sid)
print("Last two digits:", last_two_digits)
print("Number of elements for the vector:", num_elements)
a = np.arange(num_elements)
print("\n1-D vector a:")
print(a)
a_2d = a.reshape(1, -1)   
print("\n2-D array a_2d (1 row):")
print(a_2d)
b = a_2d.copy()
print("\nCopied array b:")
print(b)
print("\nShape of a:", a.shape)
print("Shape of a_2d:", a_2d.shape)
print("Shape of b:", b.shape)
```
