```python
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import iplot

df = pd.read_csv(
    r"C:\Users\sudha\Downloads\XAUUSD_1 Min_Bid_2022.01.01_2022.12.31 (2).csv"
)

start_index = 21463
time_period = 463

subset_data = df.iloc[start_index:start_index + time_period].copy()
subset_data["Time (UTC)"] = pd.to_datetime(subset_data["Time (UTC)"])

trace_high = go.Scatter(
    x=subset_data["Time (UTC)"],
    y=subset_data["High_Bid"],
    mode="lines",
    name="High_Bid"
)

trace_low = go.Scatter(
    x=subset_data["Time (UTC)"],
    y=subset_data["Low_Bid"],
    mode="lines",
    name="Low_Bid"
)

fig = go.Figure(
    data=[trace_high, trace_low],
    layout=go.Layout(
        title="High_Bid and Low_Bid Prices Over Selected Period",
        xaxis=dict(title="Time (UTC)"),
        yaxis=dict(title="Price"),
        template="plotly_white"
    )
)

iplot(fig)
```
