```python
import pandas as pd

# Step 1: n = last digit of SID (if 0, use 10)
SID = 2421463
n = SID % 10
if n == 0:
    n = 10
print("n =", n)

# Load data
df = pd.read_csv(r"C:\Users\sudha\Downloads\adult_data_mini.csv")

# Step 2: Group by relationship then hours-per-week (original)
group_original = (
    df.groupby(["relationship", "hours-per-week"])
      .size()
      .reset_index(name="count")
)
print("\nGrouped by relationship and hours-per-week (original):")
print(group_original.head(15))

# Step 3: Reduce hours-per-week by n using a FUNCTION
def reduce_hours(hours):
    return hours - n

df["reduced-hours-per-week"] = df["hours-per-week"].apply(reduce_hours)

# Step 4: Group by relationship and reduced-hours-per-week
group_reduced = (
    df.groupby(["relationship", "reduced-hours-per-week"])
      .size()
      .reset_index(name="count")
)
print("\nGrouped by relationship and reduced-hours-per-week:")
print(group_reduced.head(15))
```
